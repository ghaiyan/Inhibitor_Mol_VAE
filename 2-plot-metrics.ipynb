```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

# 读取数据
file_path = 'Inhibitor1368_data/Inhibitor1368_data_9/rnn_Inhibitor1368_data_9_data_tanimoto_similarity_screen_same.csv'
data = pd.read_csv(file_path)

# 定义函数计算分子量和LogP
def calculate_properties(smiles):
    if pd.isna(smiles):  # 检查smiles是否为NaN
        return None, None  # 如果是NaN，返回None
    mol = Chem.MolFromSmiles(str(smiles))  # 确保smiles是字符串
    if mol:
        molwt = Descriptors.MolWt(mol)
        logp = Descriptors.MolLogP(mol)
        return molwt, logp
    return None, None  # 如果分子无法从SMILES生成，返回None

# 应用函数计算每对分子的分子量和LogP
properties = data.apply(lambda row: calculate_properties(row['reconstruct_smiles']) + calculate_properties(row['raw_smiles']), axis=1, result_type='expand')
data[['reconstruct_molwt', 'reconstruct_logp', 'raw_molwt', 'raw_logp']] = properties

# 计算差值
data['delta_molwt'] = data['reconstruct_molwt'] - data['raw_molwt']
data['delta_logp'] = data['reconstruct_logp'] - data['raw_logp']

# 保存更新后的DataFrame回同一CSV文件
data.to_csv(file_path, index=False)

print("Updated data saved.")
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

# 读取数据
file_path = 'Inhibitor1368_data/Inhibitor1368_data_9/gjf/transvae/raw_smiles/caluresults.txt'
# 读取数据
data = []
with open(file_path, 'r') as file:
    # 读取表头
    header = file.readline().strip().split('\t')
    for line in file:
        id, homo_energy, lumo_energy, gap = line.strip().split('\t')
        data.append((id, float(homo_energy), float(lumo_energy), float(gap)))

# 计算硬度和软度，并更新数据
for i, (id, homo_energy, lumo_energy, gap) in enumerate(data):
    hardness = gap / 2
    softness = 1 / gap
    data[i] = (id, homo_energy, lumo_energy, gap, hardness, softness)

# 将更新后的数据写回原文件
with open(file_path, 'w') as file:
    # 写入表头
    file.write('\t'.join(header + ['Hardness', 'Softness']) + '\n')
    for row in data:
        file.write('\t'.join(map(str, row)) + '\n')

```
